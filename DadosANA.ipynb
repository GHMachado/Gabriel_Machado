# Filtros para fazer download

```python
# Listando todas as estações do hydroBR
lista_estacoes = hydrobr.get_data.ANA.list_prec_stations()
lista_estacoes.head()
```

```python
# Filtrando para o Estado e a Empresa responsável
lista_estacoes_state = lista_estacoes[lista_estacoes.State=='RIO DE JANEIRO']
lista_estacoes_state_ana = lista_estacoes_state[lista_estacoes_state.Responsible=='ANA']
lista_estacoes_state_ana
```

```python
codigos_estaçoes = lista_estacoes_state_ana.Code.to_list()[0:]
codigos_estaçoes
```

```python
dados_estacoes = hydrobr.get_data.ANA.prec_data(codigos_estaçoes)
```

```python
from plotly.offline import plot

gantt_fig = hydrobr.Plot.gantt(dados_estacoes) #Get the Gantt Fig

#Updating the layout
gantt_fig.update_layout(
    autosize=True,
    width=5000,
    height=1000,
    xaxis_title = 'Ano',
    yaxis_title = 'Código Estação',
    font=dict(family="Courier New, monospace", size=9))

#To plot and save the gantt plot as html
plot(gantt_fig,filename='gantt' + '.html') 
```
